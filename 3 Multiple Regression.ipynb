#### Import the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Read the data

```python
df = pd.read_csv('mtcars.csv')
```

```python
df
```

#### Separate the input and output variables

```python
# disp, hp and wt -> input variables
x = df[['disp','hp','wt']]

# mpg -> output variable
y = df['mpg']
```

```python
x.shape
```

#### Build the model

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x, y)
```

```python
disp = 221
hp = 102
wt = 2.91
```

```python
reg.predict([[disp,hp,wt]])
```

```python
reg.score(x, y)
```

#### Visualize

```python
plt.figure(figsize=(16,9))
plt.subplot(221)
plt.title('disp vs mpg')
plt.scatter(x['disp'], y)
plt.subplot(222)
plt.title('hp vs mpg')
plt.scatter(x['hp'], y)
plt.subplot(223)
plt.title('wt vs mpg')
plt.scatter(x['wt'], y)
```
