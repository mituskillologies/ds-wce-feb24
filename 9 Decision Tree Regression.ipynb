#### Dataset

```python
# Dataset: Petrol_Consumption.csv
# Location: https://mitu.co.in/dataset
```

#### Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read the dataset

```python
df = pd.read_csv('petrol_consume.csv')
```

```python
df
```

```python
df.corrwith(df['Petrol_Consumption'])
```

#### Separate the input and output variables

```python
x = df.drop('Petrol_Consumption', axis = 1)
y = df['Petrol_Consumption']
```

```python
x.shape
```

#### Cross Validation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

#### Build the model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(random_state=0)
```

```python
regressor.fit(x_train, y_train)
```

#### Plot the tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(16,12))
plot_tree(regressor, fontsize=7, filled=True, feature_names=x.columns);
```

#### Evaluate the model

```python
y_pred = regressor.predict(x_test)
```

```python
result = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})
```

```python
result
```

```python
np.mean(abs(y_test - y_pred))
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
imp = regressor.feature_importances_
```

```python
plt.figure(figsize=(16,9))
plt.title('Feature Importance')
plt.grid()
plt.xlabel('Feature Name')
plt.ylabel('% Importance')
plt.bar(x.columns, imp*100, color='cyan')
```
